```python
#Finding Derivative by using Function
def dy_dx(x1):
    return 2*x1
dy_dx(3)
```

```python
#Auto_grad is a core components of pytorch that provides automatic differential for Tensors
import torch
x=torch.tensor(3.0,requires_grad=True)
```

```python
y=x**2
y
```

```python

y.backward()
```

```python
x.grad
```
